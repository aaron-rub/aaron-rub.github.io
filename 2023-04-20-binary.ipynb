The decimal calcutlator is also quite simple, it has a base of 2n and prompts the user to input the exponent. This is because to find any decimals its always a base of 2 to the power of the number of bit you want. For example if you want to know what the value for the 6th bit IF it is activated, for example "1000000" The value would be 2^6. It then prompts the user for ANOTHER exponent so it can conduct basic mathmatical operations. For example "2^6 + 2^3" OR "1000000 + 1000"

```python
# Binary calculator

import math

def binary_addition(bin1, bin2):
  # convert binaries to integers
  dec1 = int(bin1, 2)
  dec2 = int(bin2, 2)
  # add integers
  dec_sum = dec1 + dec2
  # convert sum back to binary
  return bin(dec_sum)[2:]
  
def binary_subtraction(bin1, bin2):
  # convert binaries to integers
  dec1 = int(bin1, 2)
  dec2 = int(bin2, 2)
  # subtract integers
  dec_diff = dec1 - dec2
  # convert difference back to binary
  return bin(dec_diff)[2:]
  
def binary_multiplication(bin1, bin2):
  # convert binaries to integers
  dec1 = int(bin1, 2)
  dec2 = int(bin2, 2)
  # multiply integers
  dec_product = dec1 * dec2
  # convert product back to binary
  return bin(dec_product)[2:]
  
def binary_division(bin1, bin2):
  # convert binaries to integers
  dec1 = int(bin1, 2)
  dec2 = int(bin2, 2)
  # divide integers
  dec_quotient = dec1 / dec2
  # convert quotient back to binary
  return bin(int(dec_quotient))[2:]
  
def main():
  # prompt user for inputs
  bin1 = input("Enter a binary number: ")
  bin2 = input("Enter another binary number: ")
    
  # conduct operations
  print("\nBinary addition: " + binary_addition(bin1, bin2))
  print("\nBinary subtraction: " + binary_subtraction(bin1, bin2))
  print("\nBinary multiplication: " + binary_multiplication(bin1, bin2))
  print("\nBinary division: " + binary_division(bin1, bin2))



main()
```

```python
# Decimal calculator

import math

def addition(exponent1, exponent2):
  # calculate first number
  num1 = int(math.pow(2, exponent1))
  # calculate second number
  num2 = int(math.pow(2, exponent2))

  # print out calculations
  print(str(2) + "^" + str(exponent1) + " = " + str(num1) + "\n")
  print(str(2) + "^" + str(exponent2) + " = " + str(num2) + "\n")
  # calculate(addition) result of base 2^n values
  result = int(num1) + int(num2)
  # print out result
  print(str(num1) + " + " + str(num2) + " = " + str(result) + "\n")


def subtraction(exponent1, exponent2):
  # calculate first number
  num1 = int(math.pow(2, exponent1))
  # calculate second number
  num2 = int(math.pow(2, exponent2))

  # print out calculations
  print(str(2) + "^" + str(exponent1) + " = " + str(num1) + "\n")
  print(str(2) + "^" + str(exponent2) + " = " + str(num2) + "\n")
  # calculate(subtraction) result of base 2^n values
  result = int(num1) - int(num2)
  # print out result
  print(str(num1) + " - " + str(num2) + " = " + str(result) + "\n")


def multiplication(exponent1, exponent2):
  # calculate first number
  num1 = int(math.pow(2, exponent1))
  # calculate second number
  num2 = int(math.pow(2, exponent2))

  # print out calculations
  print(str(2) + "^" + str(exponent1) + " = " + str(num1) + "\n")
  print(str(2) + "^" + str(exponent2) + " = " + str(num2) + "\n")
  # calculate(multiplication) result of base 2^n values
  result = int(num1) * int(num2)
  # print out result
  print(str(num1) + " * " + str(num2) + " = " + str(result) + "\n")


def divison(exponent1, exponent2):
  # calculate first number
  num1 = int(math.pow(2, exponent1))
  # calculate second number
  num2 = int(math.pow(2, exponent2))

  print(str(2) + "^" + str(exponent1) + " = " + str(num1) + "\n")
  print(str(2) + "^" + str(exponent2) + " = " + str(num2) + "\n")
  # calculate(division) result of base 2^n values
  result = int(num1) / int(num2)
  # print out result
  print(str(num1) + " / " + str(num2) + " = " + str(result) + "\n")


# take user input
exponent1 = int(input("Enter exponent 1: "))
exponent2 = int(input("Enter exponent 2: "))

# call the functions with user input
addition(exponent1, exponent2)
subtraction(exponent1, exponent2)
multiplication(exponent1, exponent2)
divison(exponent1, exponent2)
```
